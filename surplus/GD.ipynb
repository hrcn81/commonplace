```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
X = np.array([[7],[9.5],[13],[18],[24]])
y = np.array([195,325,725,900,980])

print(X)
print(y)

plt.scatter(X,y)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X,y)

print('reg.coef_:',reg.coef_)
print('reg.intercept:',reg.intercept_)
# plt.scatter(X,y)

m = reg.coef_[0]

m
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red')
```

```python
y_pred = ((m*X) + 300)

print('\n',y_pred,'\n')

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred,color = 'green',label = 'b = 0')
plt.legend()
plt.show()

b = 300
loss_slope = -2*np.sum(y - m*X.ravel() - b)
print('\n',loss_slope,'\n')

learning_rate = 0.1
step_size = loss_slope*learning_rate
print('\n',step_size,'\n')

b = b - step_size
print('\n',b,'\n')

y_pred1 = ((m*X) + b).reshape(5)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred1,color = 'green',label = 'b = {}'.format(b))
plt.plot(X,y_pred,color = 'blue',label = 'b = 0')
plt.legend()
plt.show()
```
