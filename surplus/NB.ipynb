```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv(r"https://raw.githubusercontent.com/hrcn81/commonplace/main/Datasets/tennis_anyone.csv")
```

```python
df = data.copy()
```

```python
df = df.rename(columns = {'Play-Tennis' : 'Play-Badminton'})
```

```python
df.head()
```

```python
df_encoded = pd.get_dummies(df,dtype = int)
```

```python
df_encoded.head()
```

```python
#Create input feature and target variable
X = df_encoded.drop(columns = ('Play-Badminton_Yes'))
y = df_encoded['Play-Badminton_Yes']
```

```python
classifier = GaussianNB()
classifier.fit(X,y)
```

```python
X.head()
```

```python
y.head()
```

```python
predictions = classifier.predict(X)
```

```python
predictions
```

```python
confusion_matrix = pd.crosstab(predictions,y,rownames = ['Predicted'],colnames = ['Actual'])

confusion_matrix
```

```python
for idx, feature in enumerate(X.columns):
    print(f"\n Table for feature '{feature}':")
    if idx < len(classifier.theta_):
        print(f"Mean: {classifier.theta_[idx]}")
        print(f"Variance: {classifier.var_smoothing}")
    else:
        print("No data available for this feature")
```
